```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Loading Dataset

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

### Data Preprocessing


```python
titanic.describe()
```

```python
titanic.shape
```

```python
# Checking for null values
titanic.isnull().sum()
```

```python
mean_age = titanic['age'].mean()
titanic['age'].fillna(mean_age, inplace = True)
```

```python
titanic.isnull().sum()
```

```python
df = titanic.dropna()
df.isnull().sum()
```

### Data Visualisation

```python
# Displot
sns.displot(df['fare'])
```

```python
# Joint plot
sns.jointplot(x='fare',y='age',data=df,kind='scatter')
```

```python
# pair plot
sns.pairplot(df)
```

```python
# Rug Plot
sns.rugplot(x='age',y='fare',data=df)
```
